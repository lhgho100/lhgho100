```
# counter함수 쓰기
# 소문자 영단어 두개를 입력받는다
# 각 단어를 Counter함수를 사용해서 각 글자 빈도수 출력
# 교집합으로 단어1과 단어2의 에너그램을 찾는다.
# 에너그램을 제외한 나머지를 모은다.
# Counter함수를 이용해 찾은 글자의 빈도수로 제거할 최소갯수를 구한다.

import collections #Counter()함수 사용하기위한 모듈 불러오기

word1 = input()
word2 = input()

count1 = collections.Counter(word1) # 빈도수 계산 {딕셔너리}
count2 = collections.Counter(word2)
print(count1) #확인
print(count2) #확인
inter = count1 & count2 #변수 설정 - 교집합
print(inter) #확인
result = (count1 - inter) + (count2 - inter) #에너그램을 할 수 있는 교집합을 뺀 전체 빈도수 설정

print(sum(dict(result).values())) # 제거해야할 문자의 빈도수 values값을 더해줌 (제거해야 할 최소개수의 문자 수)
```
