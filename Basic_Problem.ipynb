```
for i in range(3): # 윳을 3번 던지겠다.
    nums = list(map(int, input().split())) # 윷의 결과를 리스트로 저장

    # split()은 입력한  0 1 0 1을 공란을 기준으로 나누어 리스트형태로 [0, 1, 0, 1]
    # map함수는 원본 리스트를 변경하지 않고 새 리스트를 생성함.

    a = nums.count(0) # 윷을 던져서 나온 배의 개수를 세고 변수 저장

    if a == 1: # 배가 1개면
        print('A') # 도 출력
    elif a == 2: # 배가 2개면
        print('B') # 개 출력
    elif a == 3: # 배가 3개면
        print('C') # 걸 출력
    elif a == 4: # 배가 4개면
        print('D') # 윷 출력
    else: # 배가 0개면
        print('E') # 모 출력
```

```
headlines = [
  "Local Bear Eaten by Man",
  "Legislature Announces New Laws",
  "Peasant Discovers Violence Inherent in System",
  "Cat Rescues Fireman Stuck in Tree",
  "Brave Knight Runs Away",
  "Papperbok Review: Totally Triffic"
]
total_str='' # headlines를 하나의 스트링으로 만들기 위한 변수 설정

for i in range(len(headlines)): # headline의 요소들을 하나씩 이어 붙여주기
    total_str += headlines[i]

print(total_str) # total_str 확인 -> headline 전체를 하나로 이어짐.

one_130='' # 130자 크기의 하나의 문자열로 이어 붙일 변수 설정

for j in range(len(total_str)): #total_str 전체 글자수로 for구문 돌림.
    one_130 += total_str[j] #한번 돌때마다 한글자씩 one_130에 더해줌
    if len(one_130) ==130: #총 길이가 130일때 출력 및 break 설정
        print(one_130)
        print(len(one_130))
        break
```
