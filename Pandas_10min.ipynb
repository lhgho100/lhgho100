```
#### 8. Reshaping (변형)
tuples=list(zip(*[['bar','bar','baz','baz',
                   'foo','foo','qux','qux'],
                  ['one','two','one','two',
                   'one','two','one','two']]))
tuples
```

```
index=pd.MultiIndex.from_tuples(tuples, names=['first','second']) ###
df=pd.DataFrame(np.random.randn(8,2),index=index, columns=['A','B']) ###
df
```

```
df2=df[:4]
df2
```

```
stacked= df2.stack() # stack() 메소드는 데이터프레임 열들의 계층을 '압축'한다.
stacked
```

```
stacked.unstack() #stack() 된것을 다시 돌리는 함수
```

```
stacked.unstack(1) #1열의 인덱스가 컬럼으로 바뀜
```

```
stacked.unstack(0) # 0열의 인덱스와 컬럼이 바뀜
```
